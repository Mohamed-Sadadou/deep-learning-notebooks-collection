```python
import torch
from diffusers import StableDiffusionPipeline

model_path = "BlackSadadou/SD-Black-IMGgen-Model-v2"
pipe = StableDiffusionPipeline.from_pretrained(model_path, torch_dtype=torch.float16)
pipe.to("cuda")

image = pipe(prompt="illustration graphique de yoda dans le style 'Black's Style'").images[0]
image.save("yoda-pokemon.png")
```
